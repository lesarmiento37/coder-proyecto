```python
import requests
import json

url = "https://dogapi.dog/api/v2//groups"
headers = {"accept": "application/json"}

response = requests.get(url, headers=headers)
response_txt = response.text
print(response_txt)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql.types import StructType, StructField, StringType, MapType

# Initialize SparkSession
spark = SparkSession.builder.appName("JSON_to_DataFrame").getOrCreate()

# Convert the JSON string to a DataFrame
df = spark.read.json(spark.sparkContext.parallelize([response_txt]), multiLine=True)

# Show the DataFrame
df.show(truncate=False)
df.printSchema()
```

```python
df.selectExpr("data[0].relationships.breeds.data").show(truncate=False)
```

```python
from pyspark.sql.functions import explode, col
exploded_df = df.select(explode(col("data")).alias("entry"))

# Extract fields from the nested structure
final_df = exploded_df.select(
    col("entry.id").alias("id"),
    col("entry.type").alias("type"),
    col("entry.attributes").alias("attributes"),
    col("entry.relationships").alias("relationships")
)

# Show the final DataFrame
final_df.show(truncate=False)
```

```python

exploded_df_2 = final_df.select(explode(col("relationships.breeds.data")).alias("entry"))
#exploded_df_2.show()
# Extract fields from the nested structure
final_df_2 = exploded_df.select(
    col("entry.id").alias("id"),
    col("entry.type").alias("type")
)

# Show the final DataFrame
final_df_2.show(truncate=False)
```

```python
SELECT * 
FROM cities
-- Inner join to countries
INNER JOIN countries
-- Match on country codes
ON cities.country_code = countries.code;
```

```python
-- Select name fields (with alias) and region 
SELECT cities.name AS city, countries.name AS country, countries.region 
FROM cities
INNER JOIN countries
ON cities.country_code = countries.code;
```

```python
-- Select fields with aliases
Select c.code as country_code,
       c.name,
       e.year,
       e.inflation_rate
FROM countries AS c
-- Join to economies (alias e)
inner Join economies as e
-- Match on code field using table aliases
on c.code = e.code;
```
